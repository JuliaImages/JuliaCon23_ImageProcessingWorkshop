```julia
using Images, TestImages, MosaicViews, ImageBinarization
```

```julia
# size = (256,256)
size = (512,512)

# original image
img = imresize(testimage("sudoku"), size)

# gray scale image
img_gray = Gray.(img)

# mosaic view
mosaic(img, img_gray; nrow =1)
```

```julia
# Let's first rotate our image
imgw = imrotate(img_gray, pi/30)
```

```julia
# Now let's zoom in a bit

imgw = parent(imgw)[65:490, 65:490]
```

```julia
# Now let's binarize an image using the `Sauvola` algorithm

alg = Sauvola(;window_size=1, bias=0.08);
img_sauvola = binarize(imgw, alg)
img_mapw = mapwindow(maximum, img_sauvola, (1, 1))
mosaicview(imgw, img_sauvola, img_mapw; ncol = 3)
```
